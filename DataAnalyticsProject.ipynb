### **CUSTOMER BEHAVIOR ANALYSIS**


```python
import pandas as pd
df=pd.read_csv('Downloads/customer_shopping_behavior.csv')

df
```

```python
df.info()
```

```python
df.describe(include="all") 
#It gives only numerical statistics by default. So add the attribute 'include'
```

```python
df.isnull().sum()
```

```python
#df['Review Rating']=df['Review Rating'].fillna(df['Review Rating'].mean())
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age group
labels = ['Young Adult','Adults','Middle Aged','Senior']
df['age_group']=pd.qcut(df['age'] , q=4 , labels=labels)
```

```python
df[['age','age_group']].head(10)
#df[['age','age_group']].head(10).sort_values(by='age',ascending=False)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']]
```

```python
# Here we can see the exact number of days between purchases for each customer.
```
